```python
country_names=final_df.Country.value_counts().index
```

```python
## pie chart -top 5 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size() ## here we can group the different features
## and check the values  rating, rating colors rating text not reted.
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index() # now we can convert this into dataframe
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# conclusion
1. when rating is between 4.5 to 4.9----->excellent
2. when rating is between 4.0 to 4.4----->verygood
3. when rating is between 3.5 to 3.9----->good
4. when rating is between 2.5 to 3.4 ----->Average
5. when rating is between 1.8 to 2.4----->poor
6. when rating is  0.0----->notrated

```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings) # we will visualize the data in barplot diagram using seaborn
```

```python
import matplotlib # here we can use matplotlib to increase the diagram size 
matplotlib.rcParams['figure.figsize']=(12,6) # here we use the rcParams parameter to increase the size of diagram
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color', data=ratings) # we will use hue:->
# rating color to use color plot in Diagram.
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color', data=ratings , palette=['white','red','orange','yellow','green','green'])## here we will mped the colours  using palette
```

```python
## count plot
sns.countplot(x="Rating color" , data=ratings, palette=['blue','red','orange','yellow','green','green'])
```
